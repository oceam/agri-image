## 関数を作るには
前述のように、新しい気温の測定値が得られるたびに平均気温の計算式を入力するのは手間がかかります。以下では、3日分の気温の数値を受け取り、その平均値を返す関数**関数** `avg_temp` を定義してみましょう。

次のような形式で、**関数定義**を記述できます。

````python
def get_avg(day_one, day_two, day_three):
    avg = (day_one + day_two + day_three) / 3
    return avg
````

Python では、**関数定義**は、上のような形をしています。
最初の行は以下のように **`def`** で始まり、最後にはコロンが必要です。

------
```Python
    def 関数名(引数, ...):
```
------
**引数**（ひきすう）とは、関数が受け取る値を指し示す変数のことです。**仮引数**（かりひきすう）ともいいます。英語ではArgumentと言うためArgと略記することがあります。

二行目、三行目に関数の処理内容が書かれています。二行目はこれまでと同様に計算していることがわかります。

三行目は、return文と呼ばれ、関数の呼び出し元に「戻り値」つまり結果を返します。`return` の後ろに空白が入ることに注意してください。

さらに、関数の二行目と三行目が字下げされていることにも注目してください。インデントと呼びます。

このインデントは飾りではなく、プログラムの構造を表す重要な役割を果たします。ここでは、どの行が関数の中にある行なのかを判別するためにインデントが必須です。今後も頻繁に使います。通常はスペース4つ（2つの場合もあり；各セル内で要統一）でインデントします。`Tab`キーを押してもスペースが入ります。

上記では、`def` の後に続く `get_avg` が関数名です。それに続く括弧の中に書かれた `day_one` と `day_two` と `day_three` は、**引数**です。また、`return` の後に来る
`avg`は**戻り値**です。

なお、上では3行で関数を書きましたが、実は2行目と3行目はまとめることができます。以下のようになります。

````python
def get_avg(day_one, day_two, day_three):
    return (day_one + day_two + day_three) / 3
````

いままで、コードを一行でも書くとその場で結果が表示されていました。しかし、いまは関数で計算式を書いたのに何も起きません。

これは、関数は計算の手順を決めているだけで、実際に計算をするためにはその関数を使う（呼び出す）必要があるためです。

では定義した関数を使ってみましょう。結果はどうなるでしょうか？

````python
get_avg(9.5, 10.1, 12.3 + 2)
````

関数呼び出しは演算式の一種なので、引数の位置には任意の式を記述できますし、関数呼び出し自体も式の中に記述できます。

````python
1.2*get_avg(9.5, 10.1, 12.3 + 2)
````

上の例では、12.3に2を足した数がまず計算され、それを使って`get_avg`が呼び出され、平均値を求めた結果が戻り、それが最後に1.2倍される、という流れがあります。

すなわち、実行中にこの式は次の段階を踏んで計算され簡略化されていくことになります。

- 段階1: `1.2*get_avg(9.5, 10.1, 12.3 + 2)`
- 段階2: `1.2*get_avg(9.5, 10.1, 14.3)`
- 段階3: `1.2*11.3`
- 段階4: `13.56`


なお、`return` は単独で（後ろに何も書かずに）使うこともできます。その場合は、「値がない」ことを意味する **`None`** という特別な値が返ります。
また、そもそも関数にreturn文が存在しない場合、自動的に`return`が最後に単独で入っているものとみなして実行されます。

### 引数名を指定して引数を与える
`linear()`の例で、引数はa, b, xという名前を持っていました。一方呼び出す側からは与えた順番にa, b, xとなっていました。これを「位置引数」と言います。Pythonでは便利な機能として、受け取る関数側での名前を指定して引数を与えることができます。この場合順番よりも名前が優先されます。これは「キーワード引数」と呼びます。

## if文による条件分岐
これまでは、常に同じ処理、同じ計算を行うプログラムを作っていました。しかし実際はそれだけではない場合も当然あります。状況によって動作を変えるために使うのが`if`です。ここでは **`if`** による**条件分岐**（**if文**）の基本を紹介します。

````python
def bmax(a,b):
    if a > b: 
        return a 
    else:
        return b
````

上の関数 `bmax` は、2つの引数の大きい方を返します。

ここで `if` による条件分岐を用いています。

----
```Python
    if a > b: 
        return a 
    else:
        return b
```
----
`a` が `b` より大きければ `a` が返され、そうでなければ、`b` が返されます。条件の後ろにはコロンが必ず付きます。

ここで、`return a` が、`if` より右にインデントされていることに注意してください。
`return a` は、`a > b` が成り立つときのみ実行されます。

**`else`** は `if` の右の条件が成り立たない場合を示しています。
`else:` として、必ず `:` が付くことに注意してください。

また、`return b` も、`else` より右にインデントされていることに注意してください。
`if` と `else` は同じインデントになります。

上記の関数bmaxの例では1行目の関数定義 `def bmax(a,b):` の後から第1レベルのインデントが開始され5行目まで続きます。これは、2-5行目はいわば1行目の「見出し」に対応する「内容」であるように、インデントで下げられた範囲は定義された関数の内容としてのプログラム文のグループです。 

さらに、3行目の一行はもう一段階インデントされています。これは2行目のif文が `True` の場合にのみ実行されるグループに属します（ここでは一行しかありませんが、複数行をこのグループに入れることもできます） 

そして、4行目の `else` ではインデントが戻されています。
5行目から再び始まる第2レベルの実行文は2行目の論理式が `False` の場合に実行されるグループに属します。 この`else`ブロックは必要なければ省略できます。

最後の行ではインデントが戻されており、これ以降は関数 `bmax()` の定義とは関係ないことがわかります。

インデントの入力は`Tab`キーで行います。消すときは通常のスペースとして消せますが、一般的に2個または4個のスペースをひとまとまりで使います。まとめて2つないし4つ消すようにしてください。

````python
bmax(3,5)
````

関数の中で `return` と式が実行されますと、関数はそこで終了となります。その後の部分は実行されません。

ですので、たとえば、上の条件分岐は以下のように書くこともできます。

----
```Python
    if a > b: 
        return a 
    return b
```
----


なお、Pythonでは、`max` という関数があらかじめ定義されています。

````python
max(3,5)
````

````python
def hd(temp, humidity):
    et = 6.1078*10**(7.5*temp/(temp+237.3))
    at = 217*et/(temp+273.15)
    return (100 - humidity) * at / 100
````
