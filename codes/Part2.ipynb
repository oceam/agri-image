# いろいろなパッケージを使って応用プログラミング

### パッケージのインポート

Pythonにはこれまでに見てきたように基本的な機能（変数、関数、条件分岐、繰り返しなど）が備わっていますが、これだけではプログラミングをするうえで十分ではありません。

一方で、すでに関数の使い方として、コードのまとまりを作って再利用できることを見てきました。実はそのようにして、すでに多数の関数が**パッケージ**として用意されています。これがPythonの大きな強みです。そうした**インポート**してから関数を呼び出して使いながら、オリジナルのプログラムを作っていくことができます。

※パッケージ内部には単に関数があるわけではなく、「クラス」や「モジュール」といった関数よりも大きな構造がありますが、非常に複雑になるためここでは踏み込みません。

パッケージには大きく二種類あります。
1. 標準ライブラリ：Pythonをインストールすると合わせてインストールされるパッケージ群で、数学計算、ファイルの取り扱い、インターネット通信の制御、その他さまざまな機能を持つものが200個程度あります。使いたいときにいつでも読み込み、すなわち**インポート**をすることができます。
2. その他のパッケージ：最初からはインストールされておらず、インターネット上で配布されるパッケージです。公式の配布サイトpypi.orgには43万件以上が登録されています。非常に特化した特殊な機能を持つようなものもあれば、汎用性の高いものとして画像の取り扱い、数値や表の操作、統計解析、その他さまざまなものがあります。これらは、インポートする前に自分でインストールする必要があります（ただし、Colabには有名なものは多数すでにインストールされています）。

### 行列演算

行列演算は線形代数の領域になります。行列の積は以下の式で計算されます。 

$C(i,j) = \sum\limits_{k=1}^{n}A(i,k)B(k,j)$  

行列 A と B の線形代数の積では、A の列数は B の行数と等しくなければなりません。 

````python
a_1=np.array(([1,1],[1,1]), dtype=np.int64)
b_1=np.array(([2,2],[2,2]), dtype=np.int64)
print("a_1:\n{}".format(a_1),'\n')
print("b_1:\n{}".format(b_1),'\n')
````

行列積を計算するには`np.matmul()`関数を使います


````python
print(np.matmul(a_1,b_1))
````

### 行列演算とは似て非なる計算
numpyでは、線形代数の行列演算ではない計算もあります。この二つは結果が異なるので注意してください。`np.multiply()`はこちらの計算で、単に対応する要素同士を掛けるだけです（左上同士、左下同士、右上同士、右下同士）。

````python
print(np.multiply(a_1,b_1))
````

**ノート：行列と多次元配列**
特に積の計算において数学的な行列として考えるか、単にリストの延長線上の数字を並べたもの（多次元配列）と考えるかは大きな違いを生みます。しかし保持するデータとしてははっきり区別されるものではなく、データだけを見て「これは行列」「これは行列ではない」と言えるわけではありません。あくまで、どのような意味で使っているのか、使う側で考える必要があるということです。

### 発展練習
時間の余った方向けです。二次元平面上のある点のxy座標(a, b)と角度thetaを引数に取り、点の座標をthetaだけ反時計回りに回転させた座標を返す関数`rotate(a, b, theta)`を作ってください。なお、`np.matmul(A, B)`の代わりに`@`演算子を用いて`A@B`で行列積を求められることも用いてください。

回転行列は以下で与えられます。これを左からかけることでベクトルを回転します。
$$R(θ)=\begin{pmatrix} \cos \theta & -\sin \theta \\ \sin \theta & \cos \theta \end{pmatrix}$$

以下がエラーなく実行できれば正しくプログラミングできていると考えられます。

````python
assert np.allclose(rotate(1, 0, 0.5*math.pi), [0, 1])
assert np.allclose(rotate(1, 0, math.pi), [-1, 0])
assert np.allclose(rotate(1, 0, 1.5*math.pi), [0, -1])
assert np.allclose(rotate(1, 0, 2*math.pi), [1, 0])
````

#### 解答例

````python
import numpy as np
import math

def rotate(a, b, theta):
  p = np.array([a, b])
  matrix = np.array([[math.cos(theta), -math.sin(theta)], [math.sin(theta), math.cos(theta)]])
  return matrix@p
````

### 行列から画像を作成
そもそも画像とは何でしょうか？ いま見ているこのモニターに表示されているものは、たくさんの格子状に配置された点（ピクセル）の集まりです。一つ一つのピクセルが明るさの値を持ち、それによって文字や画像が描かれるわけです。

さて、このように二次元の格子状に集まった数字というのは、さきほど扱った行列・多次元配列そのものです。そのため、コンピューターが画像を扱う方法として、そのように点ごとの色情報（画素値）を使うことができます。これを「ラスター画像」または「ビットマップ画像」と言います。

一つのピクセルを一つの数字（通常は0-255の範囲で大きいほど明るくなります）で表すことができるのは、モノクロ画像のみです。カラー画像を表すには、いくつか方法がありますがもっとも一般的なのは光の三原色の赤R、緑G、青Bの三つの数字（それぞれ0-255）で表すものです。「RGB画像」といった呼び方をします。この場合、画面縦横方向に対応する二次元の格子を三つ重ねたもの、すなわち奥行き方向の長さが3の三次元格子になります。これは三次元の配列として表すことができます。

なお、ラスター画像とは別に、「中心座標xy半径rの円」や「長さlの線分」といったように数学の方程式などで画像を表現する方法もあります。こちらは「ベクトル画像」と言います。ただし本講義では扱いません。



画像がどんなものかを理解したところで、さっそく行列からモノクロ画像を作ってみましょう

````python
#  9x9のゼロだけの行列
image = np.zeros((9, 9))
#  そのまま可視化
plt.imshow(image, cmap='gray')
````

````python
image
````

````python
#  0から80まで値が増えていく一次元配列
image = np.arange(81)
#  二次元に形を変える
image = image.reshape((9, 9))
#  可視化
plt.imshow(image, cmap='gray')
image
````

````python
#  ふたたびゼロのみで作った後各所の値を書き換える
image = np.zeros((9, 18)).astype(np.uint8)
# U
image[1:-3, 1] = 1
image[1:-3, 7] = 1
image[-2, 3:6] = 1
image[-3, 2] = 1
image[-3, 6] = 1
# T
image[1, 9+1:-1] = 1
image[1:-1, 9+4] = 1

plt.imshow(image,cmap='gray')
````

````python
# もう一例同様に
image2 = np.zeros((5, 5))
image2[1, 0] = 1
image2[0, 1:3] = 1
image2[1,3] = 1
image2[2, 2] = 1
image2[3, 1] = 1
image2[4, 1:4] = 1

plt.imshow(image2,cmap='gray')
````

以上では、要素の値が0か1になっています。ここでの可視化の関数は実際の最大値と最小値を見てそれを白・黒の両端に設定してくれるため、これで問題ありません。ただ、そのように自動的に配慮してくれない場合では、通常0が黒、255が白なので1はほとんど真っ黒で0と見分けがつかなくなってしまいます。

### 画像の補正

画像を分析する場合、複数の画像の間でばらつきがないように明るさを正規化（揃える）ことが必要な場合が多いです（ただし、かえって捉えたい差を消してしまう場合もあるのでケースバイケースで考える必要があります）

補正には、skimageライブラリに用意されているrescale_intensity()関数を使用します。さっそくやってみましょう。

````python
from skimage import exposure

g = photo[:,:,1] * 0.5  # 撮影時のばらつきを想定してあえていったん全体を暗くする
plt.imshow(g, cmap="gray", vmin=0, vmax=255)
````

````python
from skimage import exposure

g_adjusted = exposure.rescale_intensity(g, out_range=(0,255))
plt.imshow(g_adjusted, cmap="gray", vmin=0, vmax=255)
````

全体に暗めだったのが明るくなりました。もしこれが明るすぎる画像なら自動的に暗くなります。もちろん光の当たり方などのばらつきがなくなるわけではありませんが、たとえば毎日の画像の明るさをまとめて統一できれば役に立つことでしょう。

今回どのような補正がなされたかをもっとしっかりと可視化してみましょう。ここでは画素値のヒストグラムを描画します。Photoshopなどのソフトを使われる方はご存じかもしれませんが、暗いピクセルから明るいピクセルまで（横方向）その数の多さを積み上げ（縦方向）て表現します。

````python
plt.hist(g.flatten(), bins=30, alpha=0.5)
plt.hist(g_adjusted.flatten(), bins=30, alpha=0.5)
````

青が補正前、オレンジが補正後のヒストグラムです。補正前は左側、つまり暗いピクセルばかりが多いのに対して、補正後は右、つまり明るいピクセルもあることがわかります。
