````python
assert np.allclose(rotate(1, 0, 0.5*math.pi), [0, 1])
assert np.allclose(rotate(1, 0, math.pi), [-1, 0])
assert np.allclose(rotate(1, 0, 1.5*math.pi), [0, -1])
assert np.allclose(rotate(1, 0, 2*math.pi), [1, 0])
````

#### 

````python
import numpy as np
import math

def rotate(a, b, theta):
  p = np.array([a, b])
  matrix = np.array([[math.cos(theta), -math.sin(theta)], [math.sin(theta), math.cos(theta)]])
  return matrix@p
````
