````python
import numpy as np
import matplotlib.pyplot as plt
# Creating a larger matrix to accommodate the characters "NO"
# "NO "の文字が入るように大きめのマトリックスを作る
matrix = np.zeros((25, 50))

# Modifying the matrix to form the character "N"
# 文字 "N "を形成するためにマトリックスを修正する
matrix[2:23, 2:5] = 1   # Vertical line on the left of "N"
matrix[2:23, 20:23] = 1 # Vertical line on the right of "N"
for i in range(21):     # Diagonal line for "N"
    matrix[2 + i, 2 + i] = 1

p=15
# Modifying the matrix to form the character "O"
# 文字 "O "を形成するためにマトリックスを修正する
matrix[2:23, 20+p:23+p] = 1  # Vertical line on the left of "O"
matrix[2:23, 30+p:33+p] = 1  # Vertical line on the right of "O"
matrix[2:5, 20+p:33+p] = 1   # Top horizontal line of "O"
matrix[20:23, 20+p:33+p] = 1 # Bottom horizontal line of "O"

# Displaying the matrix as an image
plt.imshow(matrix, cmap='gray')
plt.axis('off')  # Turn off the axis
plt.show()
````

````python
# Recreating the matrix to accurately display "YES"
matrix = np.zeros((25, 60))

# Correcting the matrix to form the character "Y"
for i in range(6):  # Diagonal lines converging for "Y"
    matrix[i + 2, i + 2] = 1
    matrix[i + 2, 12 - i] = 1
matrix[8:23, 7:8] = 1  # Vertical line downwards from the center of "Y"

# Modifying the matrix to form the character "E"
matrix[2:23, 20:23] = 1  # Vertical line of "E"
matrix[2:5, 20:33] = 1   # Top horizontal line of "E"
matrix[11:14, 20:33] = 1 # Middle horizontal line of "E"
matrix[20:23, 20:33] = 1 # Bottom horizontal line of "E"

# Modifying the matrix to form the character "S"
matrix[2:5, 40:53] = 1   # Top horizontal line of "S"
matrix[2:14, 40:43] = 1  # Upper vertical line of "S"
matrix[11:14, 40:53] = 1 # Middle horizontal line of "S"
matrix[20:23, 40:53] = 1 # Bottom horizontal line of "S"
matrix[11:23, 50:53] = 1 # Lower vertical line of "S"

# Displaying the matrix as an image
plt.imshow(matrix, cmap='gray')
plt.axis('off')  # Turn off the axis
plt.show()
````
